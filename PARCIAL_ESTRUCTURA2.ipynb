```python
import pandas as pd

class Nodo:
    def __init__(self, nombre, latitud, longitud, puntozwf, propietario, contrato, sector):
        self.nombre = nombre
        self.latitud = latitud
        self.longitud = longitud
        self.propietario = propietario
        self.contrato = contrato
        self.sector = sector
        self.puntozwf = puntozwf
        self.siguiente = None
        self.anterior = None

class ListaDobleEnlazada:
    def __init__(self):
        self.cabeza = None

    def insertar(self, nombre, latitud, longitud, puntozwf, propietario, contrato, sector):
        nuevo_nodo = Nodo(nombre, latitud, longitud, puntozwf, propietario, contrato, sector)
        if not self.cabeza:
            self.cabeza = nuevo_nodo
        else:
            actual = self.cabeza
            while actual.siguiente:
                actual = actual.siguiente
            actual.siguiente = nuevo_nodo
            nuevo_nodo.anterior = actual

    def mostrar_adelante(self):
        if not self.cabeza:
            print("La lista está vacía.")
            return
        actual = self.cabeza
        while actual:
            print(f"Nombre: {actual.nombre}\nLatitud: {actual.latitud}\nLongitud: {actual.longitud}\nPuntoZonaWifi: {actual.puntozwf}\nPropietario: {actual.propietario}\nContrato: {actual.contrato}\nSector: {actual.sector}\n")
            actual = actual.siguiente
        print("None")

    def mostrar_atras(self):
        if not self.cabeza:
            print("La lista está vacía.")
            return
        actual = self.cabeza
        while actual.siguiente:
            actual = actual.siguiente
        while actual:
            print(f"Nombre: {actual.nombre}\nLatitud: {actual.latitud}\nLongitud: {actual.longitud}\nPuntoZonaWifi: {actual.puntozwf}\nPropietario: {actual.propietario}\nContrato: {actual.contrato}\nSector: {actual.sector}\n")
            actual = actual.anterior
        print("None")

    def ordenar(self, atributo):
        if not self.cabeza:
            return

        actual = self.cabeza
        while actual.siguiente:
            siguiente_nodo = actual.siguiente
            while siguiente_nodo:
                if getattr(actual, atributo) > getattr(siguiente_nodo, atributo):
                    # Intercambiar datos de los nodos
                    actual.nombre, siguiente_nodo.nombre = siguiente_nodo.nombre, actual.nombre
                    actual.latitud, siguiente_nodo.latitud = siguiente_nodo.latitud, actual.latitud
                    actual.longitud, siguiente_nodo.longitud = siguiente_nodo.longitud, actual.longitud
                    actual.puntozwf, siguiente_nodo.puntozwf = siguiente_nodo.puntozwf, actual.puntozwf
                    actual.propietario, siguiente_nodo.propietario = siguiente_nodo.propietario, actual.propietario
                    actual.contrato, siguiente_nodo.contrato = siguiente_nodo.contrato, actual.contrato
                    actual.sector, siguiente_nodo.sector = siguiente_nodo.sector, actual.sector

                siguiente_nodo = siguiente_nodo.siguiente
            actual = actual.siguiente


df_excel = pd.read_csv('Mapa_de__zonas_Wifi_del_municipio_20241019.csv')

lista_doble = ListaDobleEnlazada()

for index, row in df_excel.iterrows():
    lista_doble.insertar(row['NAME'], row['LATITUDE'], row['LONGITUDE'], row['PUNTO ZWF'], row['PROPIETARIO'], row['CONTRATO'], row['SECTOR'])

print("Lista de adelante hacia atrás:" )
lista_doble.mostrar_adelante()
print("//////////////////////////////////////////////////////////////////")
print("//////////////////////////////////////////////////////////////////")
print("//////////////////////////////////////////////////////////////////")
while True:
    respuesta = input("¿Desea insertar más datos? (S/N): ")
    if respuesta.lower() != "s":
        break

    nombre = input("Ingrese el nombre: ")
    latitud = float(input("Ingrese la latitud: "))
    longitud = float(input("Ingrese la longitud: "))
    puntozwf = input("Ingrese el PuntoZonaWifi: ")
    propietario = input("Ingrese el propietario: ")
    contrato = input("Ingrese el contrato: ")
    sector = input("Ingrese el sector: ")

    lista_doble.insertar(nombre, latitud, longitud, puntozwf, propietario, contrato, sector)

print("\nLista ordenada por nombre:")
lista_doble.ordenar('nombre')
lista_doble.mostrar_adelante()
print("//////////////////////////////////////////////////////////////////")
print("//////////////////////////////////////////////////////////////////")
print("//////////////////////////////////////////////////////////////////")

print("\nLista de atrás hacia adelante:")
lista_doble.mostrar_atras()
print("//////////////////////////////////////////////////////////////////")
print("//////////////////////////////////////////////////////////////////")
print("//////////////////////////////////////////////////////////////////")
```
